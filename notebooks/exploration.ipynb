# **Data Visualization**

```python
for i in master_df['ticker'].unique():
    df=master_df[master_df['ticker']==i]
    fig = make_subplots(
        rows=2, cols=1,
        shared_xaxes=True,
        row_heights=[0.7, 0.3],
        vertical_spacing=0.05
    )

    fig.add_trace(
        go.Candlestick(
            x=df["date"],
            open=df["open"],
            high=df["high"],
            low=df["low"],
            close=df["close"],
            name="price"
        ),
        row=1, col=1
    )

    fig.add_trace(
        go.Bar(
            x=df["date"],
            y=df["volume"],
            name="volume"
        ),
        row=2, col=1
    )

    fig.update_layout(xaxis_rangeslider_visible=False)


    fig.write_image(f"charts/png_files/{i.lower()}_ohlcv_candlestick_daily(2018-2019).png")
    fig.write_html(f"charts/html_files/{i.lower()}_ohlcv_candlestick_daily(2018-2019).html")
```
